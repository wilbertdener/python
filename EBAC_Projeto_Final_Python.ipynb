```python
df= df[['id','meses_de_relacionamento','qtd_produtos','iteracoes_12m','meses_inativo_12m','qtd_transacoes_12m']]
df.dropna()
display(df)
display(df.describe())
```

7. Etapa de análise (com visualizações);

```python
fig = sns.displot(data = df, x="iteracoes_12m")
fig.set(title="Novos clientes")
```

```python
fig = sns.displot(data = df, x="meses_inativo_12m")
fig.set(title="Total de clientes inativos")
```

```python
fig = sns.displot(data = df, x="iteracoes_12m", y="qtd_transacoes_12m")
fig.set(title="Novos clientes")
```
